```python
def newton(fxn, fxn_deriv, x0, tolerance=1e-6, max_iter=100):
    for i in range(max_iter):
        f = fxn(x0)
        fprime = fxn_deriv(x0)
        xi = x0 - f/fprime
        if abs(xi-x0) < tolerance:
            print(f"{xi} in {i} iterations")
            break
        x0 = xi
        if i == max_iter:
            print("No solution after {} iterations".format(i))
```
