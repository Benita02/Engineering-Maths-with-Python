```python
import numpy as np
import matplotlib.pyplot as plt

def fxn(x):
    return x**3 - 3*x - 4

x = np.linspace(-5, 6, 200)  # Increased number of points for smoother plot
y = fxn(x)  # Vectorized function evaluation

fig, ax = plt.subplots()
ax.plot(x, y)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')


# Customize axes
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("Plot of f(x) = x^3 - 3x - 4")
ax.grid(True)

plt.show()
```

```python
def newton(fxn, fxn_deriv, x0, tolerance=1e-6, max_iter=100):
    for i in range(max_iter):
        f = fxn(x0)
        fprime = fxn_deriv(x0)
        xi = x0 - f/fprime
        if abs(xi-x0) < tolerance:
            print(f"{xi} in {i} iterations")
            break
        x0 = xi
        if i == max_iter:
            print("No solution after {} iterations".format(i))
```
