```python
import sympy
sympy.init_printing
import numpy as np
import matplotlib.pyplot as plt
```

```python
def newton(fxn, fxn_deriv, x0, tolerance=1e-6, max_iter=20):
    for i in range(max_iter):
        f = fxn(x0)
        fprime = fxn_deriv(x0)
        xi = x0 - f/fprime
        if abs(xi-x0) < tolerance:
            print(f"{xi} in {i} iterations")
            break
        x0 = xi
        if i == max_iter:
            print("No solution after {} iterations".format(i))
    
```

#### Solve    $x^3 - 3x - 4$

```python
def fxn(x):
    return x**3 - 3*x - 4

def fxn_deriv(x):
    return 3*(x**2) - 3

x = np.linspace(-5,6,20)
y = [float(i**3-3*i-4) for i in x]
fig, ax = plt.subplots()
ax.plot(x,y)

# Set the position of the spines
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Hide the top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()


```

```python
sol = newton(fxn, fxn_deriv, 2.1, 1e-6)
sol
```

#### Solve $8 - 4.5(x - \sin{x}) = 0$

```python
def fxn(x):
    return 8 - 4.5*(x - np.sin(x))

def fxn_deriv(x):
    return 4.5*(np.cos(x) - 1)

x = np.linspace(-5,6,20)
y = [float(8 - 4.5*(i - np.sin(i))) for i in x]
fig, ax = plt.subplots()
ax.plot(x,y)

# Set the position of the spines
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Hide the top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()

```

```python
sol = newton(fxn, fxn_deriv, 5, 1e-6)
sol
```
